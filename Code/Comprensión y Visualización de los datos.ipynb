```python
#Visualizacion de histogramas
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Cargar el archivo CSV
data = pd.read_csv('D:\FDS\GBvideos_cc50_202101.csv')  # Reemplaza con la ruta correcta de tu archivo

# Configuración del estilo de los gráficos
sns.set(style="whitegrid")

# Crear figuras y ejes
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Títulos de los gráficos
titles = ['Distribución de Vistas', 'Distribución de Likes', 'Distribución de Dislikes', 'Distribución de Comentarios']

# Lista de columnas a graficar
columns = ['views', 'likes', 'dislikes', 'comment_count']

# Bucle para crear cada histograma
for ax, col, title in zip(axs.flatten(), columns, titles):
    sns.histplot(data[col], ax=ax, kde=True, color='skyblue')
    ax.set_title(title)
    ax.set_xlabel(col.capitalize())
    ax.set_ylabel('Frecuencia')
    ax.set_xscale('log')  # Escala logarítmica para manejar el sesgo en los datos

# Ajuste de la disposición de los gráficos
plt.tight_layout()

plt.show()
```

```python
#Visualizacion de dispercion
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Cargar el archivo CSV
data = pd.read_csv('D:\FDS\GBvideos_cc50_202101.csv')  

# Crear un gráfico de dispersión para la relación entre vistas y likes/dislikes
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Relación entre vistas y likes
sns.scatterplot(x='views', y='likes', data=data, ax=axs[0], color='green')
axs[0].set_title('Relación entre Vistas y Likes')
axs[0].set_xscale('log')
axs[0].set_yscale('log')
axs[0].set_xlabel('Vistas (log)')
axs[0].set_ylabel('Likes (log)')

# Relación entre vistas y dislikes
sns.scatterplot(x='views', y='dislikes', data=data, ax=axs[1], color='red')
axs[1].set_title('Relación entre Vistas y Dislikes')
axs[1].set_xscale('log')
axs[1].set_yscale('log')
axs[1].set_xlabel('Vistas (log)')
axs[1].set_ylabel('Dislikes (log)')

plt.tight_layout()
plt.show()
```

```python
#Verificacion de la calidad de los datos
import pandas as pd

# Cargar el archivo CSV
data = pd.read_csv('D:\FDS\GBvideos_cc50_202101.csv')  

# 1. Verificar la Consistencia de los Datos
# Comprobación de los tipos de datos
tipos_datos = data.dtypes

# 2. Identificar y Analizar los Valores Nulos
valores_nulos = data.isnull().sum()
proporcion_nulos = valores_nulos / len(data)

# 3. Detectar Valores Atípicos
# Puedes usar métodos como IQR (Rango Intercuartílico) para identificar valores atípicos en columnas numéricas
from scipy import stats
import numpy as np

# Calculando el IQR para la columna 'views' como ejemplo
Q1 = data['views'].quantile(0.25)
Q3 = data['views'].quantile(0.75)
IQR = Q3 - Q1
outliers = ((data['views'] < (Q1 - 1.5 * IQR)) | (data['views'] > (Q3 + 1.5 * IQR)))

# Resumen de la verificación de calidad de los datos
resumen_calidad = {
    "Tipos de Datos": tipos_datos,
    "Valores Nulos": valores_nulos,
    "Proporción de Nulos": proporcion_nulos,
    "Outliers en 'views'": outliers.value_counts()
}

# Muestra el resumen
for key, value in resumen_calidad.items():
    print(f"{key}:\n{value}\n")
```
